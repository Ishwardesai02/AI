```python
import numpy as np
import cv2
import  matplotlib.pyplot as plt
%matplotlib inline
```

```python
#PUT All in one cell

#Connects to your computer default camera
cap=cv2.VideoCapture(0)

#Automatically grab width and height from the video feed
#(returns float which we need to convert to integer for later on)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
heigt=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while True:
    
    #capture frame-by-frame
    ret,frame=cap.read()
    
    #Our operation on the frame come here
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    #Display the resulting frame
    cv2.imshow('frame',gray)
    
    #This command lets us quit with the "q" button on a  keybaoard
    #Simply pressing X on the window won't work1
    if cv2.waitKey(1)&0xFF == ord('q'):
        break
        
#When everything done ,release the capture and destroy the windows
cap.release()
cv2.destroyAllWindows()
    
```

```python
#Connects to your Computer's camera
cap=cv2.VideoCapture(0)



#Automatically grab width and height from video feed
#(returns float which we need to covert to integer for later on!)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#MacOs and linux :*'XVID'
#Windows:*'VIDX'
writer=cv2.VideoWriter("C:/Users/ADMIN/6-Computer Vision/stue_cap.mp4",cv2.VideoWriter_fourcc(*'XVID'),25,(width,height))

#This loop keep recording until you hit 0 or escape the window
while True:
    #capture frame by frame
    ret,frame=cap.read()
    
    #write the video
    writer.write(frame)
    
    #Display the resulting frame
    cv2.imshow('frame',frame)
    
    #This command let's us quit with "q" button on a keyboard.
    #Simply pressing X on the window won't work
    if cv2.waitKey(1) & 0xff==ord('q'):
        break
        
#When everything done,release the capture and destroy the windows
cap.release()
writer.release()
cv2.destroyAllWindows()

    
```

```python
import cv2
import time
#same command function as streaming ,its just now we passssed in the file path ,
cap=cv2.VideoCapture('C:/Users/ADMIN/6-Computer Vision/stue_cap.mp4')

#Frames per sec for video
fps=25

#Always a good idea to check if the video was accurately there 
#If you get an error at this step ,triple check your file path

if cap.isOpened()==False:
    print('Error openiing the video file')
    
#while the video is opened
while cap.isOpened():
    #read the video file
    ret,frame=cap.read()
    
    #if we got frames
    if ret==True:
        
        #Display the frame at some frame rate of recording
        time.sleep(1/fps)
        cv2.imshow('frame',frame)
        
        #press q to quit
        if cv2.waitKey(1) & 0xff==ord('q'):
            break
    
    #OR Automatically break this while loop if the video is over
    else:
        break
        
cap.release()
#Close all the frames
cv2.destroyAllWindows()
        
    
    
```
